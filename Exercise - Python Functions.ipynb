```python
#Task 3: Function with Multiple Parameters

def calculate_area(length, width):
    return length * width
print(calculate_area(9.4, 3.5))
```

```python
#Task 4: Function with Default Parameter Values

def power(base, exponent=2):
    return base ** exponent
print(power(25, exponent=2))
```

```python
#Task 5: Function that Returns a Value

def get_full_name(first, last, middle=""):
    if middle:
        return f"{first} {middle} {last}"
    return f"{first} {last}"
print(get_full_name("Peter", "Sharma", "M"))
print(get_full_name("Peter", "Sharma"))
```

```python
#Task 6: Function Composition

def celsius_to_fahrenhit(celsius):
    return celsius * 9/5 + 32
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9
print(celsius_to_fahrenhit(25))
print(fahrenheit_to_celsius(68)) 
```

```python
#Task 7: Combining Functions and Conditionals
def analyze_temperature(value, unit):
    unit = unit.upper() 
    if unit == 'F':
        temp_c = (value - 32) * 5/9
        converted = f"{temp_c:.1f}°C"
        check_val = temp_c
    else:
        temp_f = (value * 9/5) + 32
        converted = f"{temp_f:.1f}°F"
        check_val = value
        
    if check_val < 0:
        desc = "freezing"
    elif check_val <= 15:
        desc = "cold"
    elif check_val <= 25:
        desc = "moderate"
    else:
        desc = "hot"
        
    return f"Temperature: {converted}, The weather is {desc}."

print(analyze_temperature(56, 'F'))
print(analyze_temperature(26, 'C'))
```

```python
#Task 8: Function with Variable Arguments

def calculate_average(*numbers):
    return sum(numbers)/len(numbers) if numbers else 0
print(calculate_average(3, 5, 17, 49, 56, 98))
```
